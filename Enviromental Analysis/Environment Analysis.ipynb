```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("Manesar_Data.csv")
df.head()
```

```python
print(df.info())

print(df.describe())

print(df.isnull().sum())
```

```python
plt.figure(figsize=(6, 4))
sns.histplot(df['AQI'].dropna(), bins=10, kde=True)
plt.title("Air Quality Index (AQI) Distribution")
plt.xlabel("AQI OF MANESAR")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(8, 4))
sns.barplot(x='Soil Type', y='Water Pollution Index', data=df)
plt.title("Average Water Pollution Index by Soil Type")
plt.xlabel("Soil Type of manesar")
plt.ylabel("Water Pollution Index")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(18, 14))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='YlGnBu')
plt.title("Correlation Between Environmental Factors")
plt.show()
```
